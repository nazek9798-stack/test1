```python
# Convert 'date' column to datetime objects
df['date'] = pd.to_datetime(df['date'])

# Filter data for November (assuming the current year is 2023 based on the sample data)
november_mau_df = df[(df['date'].dt.month == 11) & (df['date'].dt.year == 2023)]

# Calculate MAU by counting unique user_id's in November
mau_november = november_mau_df['user_id'].nunique()

print(f"Monthly Active Users (MAU) for November 2023: {mau_november}")
```

```python
# Calculate DAU by grouping by date and counting unique user_id's
dau_df = df.groupby(df['date'].dt.date)['user_id'].nunique().reset_index()
dau_df.columns = ['Date', 'DAU']

print("Daily Active Users (DAU) for November 2023:")
display(dau_df)

# Optionally, calculate average DAU for November
average_dau_november = dau_df['DAU'].mean()
print(f"\nAverage Daily Active Users (DAU) for November 2023: {average_dau_november:.2f}")
```

```python
# Get unique users who appeared on November 1st, 2023
nov1_users = df[df['date'].dt.date == pd.to_datetime('2023-11-01').date()]['user_id'].unique()

# Get unique users who appeared on November 2nd, 2023
nov2_users = df[df['date'].dt.date == pd.to_datetime('2023-11-02').date()]['user_id'].unique()

# Calculate the number of users who returned on day 2
returned_users_day1_retention = len(set(nov1_users) & set(nov2_users))

# Calculate the retention rate for day 1
retention_day1 = (returned_users_day1_retention / len(nov1_users)) * 100

print(f"Users who first joined on November 1st, 2023: {len(nov1_users)}")
print(f"Users from November 1st who returned on November 2nd: {returned_users_day1_retention}")
print(f"1-day Retention Rate for November 1st joiners: {retention_day1:.2f}%")
```

```python
# Identify unique users who viewed at least one advert in November
users_viewed_adverts_november = november_mau_df[november_mau_df['view_adverts'] > 0]['user_id'].nunique()

# Calculate the user conversion rate
conversion_rate_adverts = (users_viewed_adverts_november / mau_november) * 100

print(f"Total unique users in November (MAU): {mau_november}")
print(f"Unique users who viewed at least one advert in November: {users_viewed_adverts_november}")
print(f"User conversion rate for viewing adverts in November: {conversion_rate_adverts:.2f}%")
```

```python
total_ad_views_november = november_mau_df['view_adverts'].sum()
average_ad_views_per_user = total_ad_views_november / mau_november

print(f"Общее количество просмотренных объявлений в ноябре: {total_ad_views_november}")
print(f"Среднее количество просмотренных объявлений на пользователя в ноябре: {average_ad_views_per_user:.2f}")
```

```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/test.xlsx', sheet_name='Данные АБ тестов')
df.head()
```

```python
from scipy import stats

# Get unique experiment numbers
experiment_numbers = df['experiment_num'].unique()

for exp_num in sorted(experiment_numbers):
    print(f"\n--- Анализ эксперимента №{exp_num} ---")

    # Filter data for the current experiment
    experiment_df = df[df['experiment_num'] == exp_num]

    # Separate control and test groups
    control_group = experiment_df[experiment_df['experiment_group'] == 'control']
    test_group = experiment_df[experiment_df['experiment_group'] == 'test']

    # Calculate ARPU for both groups
    arpu_control = control_group['revenue'].sum() / control_group['user_id'].nunique()
    arpu_test = test_group['revenue'].sum() / test_group['user_id'].nunique()

    print(f"  ARPU контрольной группы: {arpu_control:.2f}")
    print(f"  ARPU тестовой группы: {arpu_test:.2f}")

    # Perform Welch's t-test (handles unequal variances and sample sizes)
    # We need the revenue values for each user to perform the test, not just the ARPU.
    # For ARPU comparison, we are comparing the means of revenue per user.
    # Since the revenue column contains individual user revenues, we can directly compare these distributions.
    # For a more precise ARPU comparison with a t-test, we would ideally need a single revenue per user for each user, then compare those distributions.
    # However, if 'revenue' represents each transaction, we should sum revenues per user first.

    # Let's assume 'revenue' column already represents the total revenue per user for simplicity as per the context description.
    # If 'revenue' can have multiple entries for the same user, we should aggregate it first.
    # Given 'revenue - выручка, которую сгенерировал пользователь, купив платную услугу продвижения', it implies total revenue per user.

    # To be safe, let's group by user_id and sum revenue to get actual revenue per unique user.
    control_revenue_per_user = control_group.groupby('user_id')['revenue'].sum()
    test_revenue_per_user = test_group.groupby('user_id')['revenue'].sum()

    # Perform independent t-test (Welch's t-test, assuming unequal variances)
    t_stat, p_value = stats.ttest_ind(test_revenue_per_user, control_revenue_per_user, equal_var=False)

    print(f"  P-значение (Welch's t-test): {p_value:.4f}")

    alpha = 0.05

    if p_value < alpha:
        print("  Вывод: Существуют статистически значимые различия между группами.")
        if arpu_test > arpu_control:
            print("  Рекомендация: Тестовая версия показала более высокий ARPU. Рекомендуется внедрить изменения.")
        else:
            print("  Рекомендация: Контрольная версия показала более высокий ARPU. Не рекомендуется внедрять изменения, или требуется дополнительный анализ.")
    else:
        print("  Вывод: Статистически значимых различий между группами нет.")
        print("  Рекомендация: Изменения не оказали статистически значимого влияния на ARPU. Дополнительный анализ или дальнейшие тесты могут быть полезны.")
```

```python
# Calculate total revenue from the 'revenue' column
total_revenue = df['revenue'].sum()

# Calculate total unique users from the 'user_id' column
total_unique_users = df['user_id'].nunique()

# Calculate Average Revenue Per User (ARPU)
overall_arpu = total_revenue / total_unique_users

print(f"Средний доход на пользователя (ARPU) по датасету 'Листеры': {overall_arpu:.2f}")
```
